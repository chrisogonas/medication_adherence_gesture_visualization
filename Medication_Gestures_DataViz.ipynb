# Medication Adherence Study - Gesture Data Analysis

### Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
from matplotlib import colors as mcolors
from matplotlib import style
```

## Part 1 - Data Preprocessing

```
df = pd.read_csv('/content/sample_data/med_data/raw_data_apple.csv')
df_single = pd.read_csv('/content/sample_data/med_data/raw_data_single_ges.csv')
df1 = pd.read_csv('/content/sample_data/med_data/raw_data_single_ges_open_bottle.csv')
df2 = pd.read_csv('/content/sample_data/med_data/raw_data_single_ges_hand2mouth.csv')
df3 = pd.read_csv('/content/sample_data/med_data/raw_data_single_ges_pick_drink_return.csv')
```

```
df_single.head()
```

## Part 2 - Data Visualization

### Scatter Plot - 3 gestures

```
sns.set(style = "darkgrid")
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111, projection = '3d')

for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
  x = df['x']
  y = df['y']
  z = df['z']
  ax.scatter(x, y, z, c=c, marker=m)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

### Scatter Plot - 1 gesture

```
sns.set(style = "darkgrid")
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111, projection = '3d')

for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
  x = df_single['x']
  y = df_single['y']
  z = df_single['z']
  ax.scatter(x, y, z, c=c, marker=m)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

### Line Plot - 3 gestures

```
plt.rcParams["figure.figsize"] = (20,10)
style.use('ggplot')
x = df['x']
y = df['y']
z = df['z']
cd = [4,14,8]
plt.plot(x,label='x values', linewidth=3,color='red')
plt.plot(y,label='y values',linewidth=3,color='blue')
plt.plot(z,label='z values',linewidth=3,color='orange')
plt.title('Medication Hand Motion Gestures (3 gestures)')
plt.ylabel('Y coordinate')
plt.xlabel('x coordinate')
plt.legend()
plt.show()
```
