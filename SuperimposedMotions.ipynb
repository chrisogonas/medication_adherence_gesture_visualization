# Medication Adherence Hand Gesture Accelerometer Data Superimposed

## Import Libraries

```
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
from matplotlib import colors as mcolors
from matplotlib import style
```

## Load Data

```
df = pd.read_csv('/content/sample_data/mtpdata/user1_and.csv')
df1 = pd.read_csv('/content/sample_data/mtpdata/user2_cin.csv')
df2 = pd.read_csv('/content/sample_data/mtpdata/user3_coo.csv')
```

```
df.head()
```

## Visualize Data

### User 1

```
plt.rcParams["figure.figsize"] = (20,10)
style.use('ggplot')
x1 = df['x1']
x2 = df['x2']
x3 = df['x3']
y1 = df['y1']
y2 = df['y2']
y3 = df['y3']
z1 = df['z1']
z2 = df['z2']
z3 = df['z3']

cd = [4,14,8]
plt.plot(x1,label='x1 values', linewidth=3,color='red')
plt.plot(x2,label='x2 values', linewidth=3,color='red')
plt.plot(x3,label='x3 values', linewidth=3,color='red')
plt.plot(y1,label='y1 values',linewidth=3,color='blue')
plt.plot(y2,label='y2 values',linewidth=3,color='blue')
plt.plot(y3,label='y3 values',linewidth=3,color='blue')
plt.plot(z1,label='z1 values',linewidth=3,color='orange')
plt.plot(z2,label='z2 values',linewidth=3,color='orange')
plt.plot(z3,label='z3 values',linewidth=3,color='orange')
plt.title('Medication Hand Motion Gestures - 3 gestures superimposed')
plt.ylabel('Y')
plt.xlabel('x')
plt.legend()
plt.show()
```
